```python
X = df.drop(['animal_name','class_type'], axis=1)
y = df['class_type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size  = 0.2, random_state =0 )
```

# Model

```python
model = Sequential()
model.add(InputLayer(input_shape=(16,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(7, activation='softmax'))
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(16,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(7, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Convert integer labels to one-hot encoded labels
# Subtract 1 from y_train/y_test because class labels are 1-indexed (1 to 7), and to_categorical expects 0-indexed.
y_train_ohe = to_categorical(y_train - 1, num_classes=y.max())
y_test_ohe = to_categorical(y_test - 1, num_classes=y.max())

model.fit(X_train, y_train_ohe, epochs=100, batch_size=32, validation_data=(X_test, y_test_ohe))
```

```python
model.predict(X_test)
```

```python
pred = (model.predict(X_test) > 0.5).astype(int)[:,0]
pred
```

```python
print(classification_report(y_test, pred))
```
